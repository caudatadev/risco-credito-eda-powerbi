##  Trocando o tipo da coluna para poder alterar 1 e 0 para inadimplente e não inadimplente

```python
dados["inadimplencia"] = dados["inadimplencia"].astype(str)
```

```python
traducao1 = {
            "RENT": "Aluguel",
            "OWN":"Própria",
            "MORTGAGE":"Financiada",
            "OTHER":"Outros"
}
dados["situacao_moradia"] = dados["situacao_moradia"].map(traducao1)

traducao2 = {
        'PERSONAL':"Pessoal",
        'EDUCATION':"Educação", 
        'MEDICAL':"Saúde", 
        'VENTURE':"Empreendimento", 
        'HOMEIMPROVEMENT':"Reforma",
        'DEBTCONSOLIDATION': "Consolidação de Dividas" 
}
dados["motivo_emprestimo"] = dados["motivo_emprestimo"].map(traducao2)

```

```python
traducao3 = {
    "1":"Inadimplente",
    "0":"Não Inadimplente"    
}

dados["inadimplencia"] = dados["inadimplencia"].map(traducao3)
```

```python
traducao4 = {
    "Y":"S",
    "N":"N"
}

dados["historico_inadimplencia"] = dados["historico_inadimplencia"].map(traducao4)
```

```python
dados.head()
```

    Ajustando registros

    Preenchendo os dados nulos da coluna tempo_emprego_anos com a mediana (decidi preencher dessa maneira por causa de alguns possiveis valores discrepantes)

```python
#Arrumando um erro de digitação no primeiro registro da coluna tempo_emprego_anos
dados.iloc[0,3] = 4.0
```

```python
dados["tempo_emprego_anos"] = dados["tempo_emprego_anos"].fillna(dados["tempo_emprego_anos"].median())
```

```python
dados.head()
```

```python
#Mudando o tipo de dado da coluna até por que não faz sentido uma coluna de anos ser um float
dados["tempo_emprego_anos"] = dados["tempo_emprego_anos"].astype("int64")
```

```python
porcentagem_nulos = round(dados.isnull().mean() * 100,2)

print(f"Porcentagem de dados nulos:",porcentagem_nulos["taxa_juros"])

```

```python
dados.head()
```

> **💡 Decisão de Negócio (Tratamento de Nulos):** Em vez de utilizar a média global e distorcer a regra de precificação da instituição financeira, os 9% de dados faltantes na `taxa_juros` foram preenchidos utilizando a média de agrupamentos baseados no risco do cliente (Cruzamento de Inadimplência Atual x Histórico de Inadimplência). Isso preserva a variância natural onde clientes mais arriscados recebem taxas maiores.

```python
grupo_preenche_nulo = dados.groupby(["inadimplencia","historico_inadimplencia"])["taxa_juros"].transform("mean")

dados["taxa_juros"] = dados["taxa_juros"].fillna(grupo_preenche_nulo)

dados["taxa_juros"] = round(dados["taxa_juros"],2)


```

```python
dados.isnull().sum()
```

```python
print("Porcentagem de nulos na Taxa de Juros após o preenchimento:")
print(dados['taxa_juros'].isnull().mean() * 100)
```

```python
dados["renda_mensal"] = round(dados["renda_anual"] / 12,0)

limites_idade = [0, 25, 35, 45, 60, 120]
nomes_faixas = ['Até 25 anos', '26-35 anos', '36-45 anos', '46-60 anos', 'Mais de 60 anos']
dados['faixa_etaria'] = pd.cut(dados['idade'], bins=limites_idade, labels=nomes_faixas)
```

```python
dados.info()
```

```python
dados["renda_mensal"] = dados["renda_mensal"].astype("Int64")
```

Identificando quais são os outliers que estão gerando 5 nulos na coluna faixa etaria

```python
linhas_com_nulo = dados[dados["faixa_etaria"].isnull()]

print(linhas_com_nulo)
```

Durante a checagem das novas colunas, identifiquei 5 registros com idades irreais (entre 123 e 144 anos), caracterizando erros de digitação ou falhas no sistema legado. Como representavam uma parcela ínfima da base, optei pela exclusão (drop) dessas linhas para não enviesar as faixas etárias.
